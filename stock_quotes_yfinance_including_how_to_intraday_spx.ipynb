```
stock = 'SPY'
mydf = yf.Ticker(stock)
expirations = mydf.options

all_options_dfs = []

for expiration in expirations:
  opts = mydf.option_chain(expiration)
  c = opts.calls
  p = opts.puts
  all_options_dfs.append(c)
all_options_dfs  
```
